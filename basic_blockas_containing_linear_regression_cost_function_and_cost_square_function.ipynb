```python
import numpy as np # We import numpy to handle our arrays of data efficiently

# Our training data
x_train = np.array([1.0, 2.0]) # House 1 is 1000 sq ft. House 2 is 2000 sq ft.
y_train = np.array([300.0, 500.0]) # House 1 costs $300k. House 2 costs $500k.

# m is the number of training examples (in this case, m = 2)
m = x_train.shape[0]
print(f"Number of training examples (m): {m}")
```

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model.

    Parameters:
      x (ndarray): This is our input data (e.g., house sizes). It's an array of shape (m,).
      w (scalar):  This is the Weight (slope of the line).
      b (scalar):  This is the Bias (y-intercept).

    Returns:
      y_hat (ndarray): These are the model's predictions, shape (m,).
    """
    # 1. Find out how many examples we have by checking the length of the x array
    m = x.shape[0]

    # 2. Create an empty array of the same size to store our predictions
    # np.zeros simply creates an array filled with 0.0s, like: [0., 0.]
    y_hat = np.zeros(m)

    # 3. Loop through every single data point in our dataset
    for i in range(m):
        # 4. Apply the mathematical formula: f_w,b(x) = w * x + b
        # We grab the i-th input value, multiply by w, and add b.
        y_hat[i] = w * x[i] + b

    # 5. Return the array of calculated predictions back to the user
    return y_hat

# --- Let's test it out! ---
# Let's guess that the weight is 200 (price goes up $200k per 1000 sq ft)
# Let's guess the bias is 100 (base price is $100k)
w_guess = 200.0
b_guess = 100.0

# Call our function
predictions = compute_model_output(x_train, w_guess, b_guess)
print(f"Our model predicted: {predictions}")
# Look at the output: For x=1.0, 200(1)+100 = 300. For x=2.0, 200(2)+100 = 500.
```

```python
def compute_cost(x, y, w, b):
    """
    Computes the Mean Squared Error cost.

    Parameters:
      x (ndarray): Input features, shape (m,)
      y (ndarray): Target true values, shape (m,)
      w, b (scalar): Model parameters

    Returns:
      total_cost (float): The cost of using w,b to predict y
    """
    # 1. Get the number of training examples
    m = x.shape[0]

    # 2. Initialize a variable to keep a running total of our errors
    sum_of_squared_errors = 0

    # 3. Loop through every single data point
    for i in range(m):
        # 4. Calculate what the model predicts for this specific x[i]
        prediction = w * x[i] + b

        # 5. Calculate the raw error (Prediction - Actual Truth)
        error = prediction - y[i]

        # 6. Square the error (this makes it positive and amplifies big mistakes)
        squared_error = error ** 2

        # 7. Add this squared error to our running total
        sum_of_squared_errors = sum_of_squared_errors + squared_error

    # 8. Divide the total sum by 2m to get the final average cost
    total_cost = (1 / (2 * m)) * sum_of_squared_errors

    # 9. Return the final number
    return total_cost
```
