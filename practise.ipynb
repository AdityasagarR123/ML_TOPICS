```python
import numpy as np

# 1. Generate Synthetic 1D Data
np.random.seed(42)
# Create 100 random values between 0 and 2
X = 2 * np.random.rand(100)
# True relationship: y = 3x + 4, plus some random Gaussian noise
y = 3 * X + 4 + np.random.randn(100) * 0.5

# 2. Define Core Functions
def predict(x, w, b):
    return w * x + b

def compute_cost(x, y, w, b):
    m = x.shape[0]
    predictions = predict(x, w, b)
    squared_errors = (predictions - y) ** 2
    return (1 / (2 * m)) * np.sum(squared_errors)

def compute_gradient(x, y, w, b):
    m = x.shape[0]
    predictions = predict(x, w, b)
    errors = predictions - y
    dj_dw = (1 / m) * np.sum(errors * x)
    dj_db = (1 / m) * np.sum(errors)
    return dj_dw, dj_db

# 3. Gradient Descent Loop
def gradient_descent(x, y, w_init, b_init, alpha, num_iters):
    w = w_init
    b = b_init

    print(f"Starting Gradient Descent...")
    print(f"Initial w: {w:.4f}, b: {b:.4f}, Cost: {compute_cost(x, y, w, b):.4f}\n")

    for i in range(num_iters):
        # Calculate gradients
        dj_dw, dj_db = compute_gradient(x, y, w, b)

        # Update parameters simultaneously
        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        # Print progress every 100 iterations
        if i % 100 == 0 or i == num_iters - 1:
            cost = compute_cost(x, y, w, b)
            print(f"Iteration {i:4d}: Cost = {cost:.4f}, w = {w:.4f}, b = {b:.4f}")

    return w, b

# 4. Execute
# Initialize parameters to 0
w_initial = 0.0
b_initial = 0.0

# Hyperparameters
learning_rate = 0.1
iterations = 1000

w_final, b_final = gradient_descent(X, y, w_initial, b_initial, learning_rate, iterations)

print("\n--- Final Results ---")
print(f"True parameters : w = 3.0000, b = 4.0000")
print(f"Learned params  : w = {w_final:.4f}, b = {b_final:.4f}")
```
