```python
import numpy as np
import matplotlib.pyplot as plt

# 1. The Raw Data
# Let's say this is our target data: House sizes vs Prices
x = np.array([1.0, 2.0, 3.0])
y_true = np.array([250.0, 450.0, 650.0])

# 2. Initialize the canvas
plt.figure(figsize=(10, 6))

# Plot the actual true data points in black
plt.scatter(x, y_true, color='black', s=100, label='Actual True Data', zorder=5)

# 3. Case 1: High Weight, Zero Bias (w = 300, b = 0)
# Notice how this line starts at 0 and shoots up way too fast.
w1, b1 = 300.0, 0.0
y_hat1 = w1 * x + b1
plt.plot(x, y_hat1, color='red', linestyle='--', label=f'Case 1: w={w1}, b={b1} (Too Steep)')

# 4. Case 2: Low Weight, High Bias (w = 50, b = 400)
# Notice how this line starts way too high, and doesn't tilt up fast enough.
w2, b2 = 50.0, 400.0
y_hat2 = w2 * x + b2
plt.plot(x, y_hat2, color='blue', linestyle='--', label=f'Case 2: w={w2}, b={b2} (Too Flat, Too High)')

# 5. Case 3: The "Perfect" Fit (w = 200, b = 50)
# This is what Gradient Descent is trying to find.
w3, b3 = 200.0, 50.0
y_hat3 = w3 * x + b3
plt.plot(x, y_hat3, color='green', linewidth=3, label=f'Case 3: w={w3}, b={b3} (Perfect Fit)')

# 6. Formatting the diagram to make it readable
plt.title("Visualizing Weights (w) and Biases (b)", fontsize=16)
plt.xlabel("House Size (1000s of sq ft)", fontsize=12)
plt.ylabel("Price ($1000s)", fontsize=12)
plt.axhline(0, color='black',linewidth=1) # x-axis
plt.axvline(0, color='black',linewidth=1) # y-axis
plt.xlim(0, 3.5)
plt.ylim(0, 1000)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

# Render the diagram
plt.show()
```
